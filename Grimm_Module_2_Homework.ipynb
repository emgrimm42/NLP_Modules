3. Clean, tokenize, and lemmatize the corpus

3.1. Clean the dataset, aka remove noise like punctuation and caapatalization to make the dataset more easily readable for the computer

```python

import re
#Function to clean

def initial_clean(text):
    text = re.sub(r'[^\w\s]','',text)
    text = text.lower()
    return text

df['cleaned'] = df['plainpagefulltext'].apply(initial_clean)
df['cleaned'][:5]
```

Seems to have worked, so let's proceed to step:  
3.2. Tokenization, aka breaking the text down to smaller, easier to process bits, or tokens, usually individual words or even word-bits. This is fundamental in NLP, because it allows the computer to actually process and analyse the words.


```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

def tokenize(text):
  text = nltk.word_tokenize(text, language = 'german')
  return text

#continue the code
df['tokenized'] = df['cleaned'].apply(tokenize)
df['tokenized'][:5]
```

The different texts are now represented by lists containing all the same words in the same order as opposed to one long string. Next is:
3.3. Lemmatization, or returning all the words to their word stem so that the same word but written in two different tenses isn't considered one word.

```python
import spacy
from nltk.corpus import stopwords
nltk.download('stopwords')



# Ensure the German language model is downloaded
!python -m spacy download de_core_news_sm

# Load the German language model
nlp = spacy.load('de_core_news_sm')

# Get German stop words and add custom ones
stop_words = set(stopwords.words('german'))
# why are the custom stop words in a dictionary not a list?
custom_stop_words = {'herr', 'frau', 'dez', 'januar', 'ge', 'nr', 'ſind', 'handeln', 'ver', 'ent', 'bi'}
stop_words.update(custom_stop_words)

def lemmatize_and_remove_stopwords(texts):
    texts_out = []
    for sent in texts:
        # Join the tokens into a single string
        text = " ".join(sent)


        # Process the text with spaCy
        doc = nlp(text)

        # Lemmatize, lowercase, and remove stop words
        lemmatized = [token.lemma_.lower() for token in doc
                      if token.is_alpha and token.lemma_.lower() not in stop_words]

        texts_out.append(lemmatized)

    return texts_out

# Apply the function to the DataFrame
df['lemmatized'] = lemmatize_and_remove_stopwords(df['tokenized'])

# Display the first 5 rows of the lemmatized column
df['lemmatized'].head()
```

4. Is to apply the NLP package to sort out the verbs from the data.  
For this we need parts of speech tagging (POS tagging) which we did in the data detective part of unit 2, so we can just copy out the code from there and alter it slightly so that it sorts out all the verbs instead of all the nouns.

```python
def tagging(texts, allowed_postags=['VERB']): # possible tags'NOUN', 'ADJ', 'ADV', 'VERB'
    texts_out = []
    nlp = spacy.load('de_core_news_sm')
    for sent in texts:
        sent_str = " ".join(sent)
        doc = nlp(sent_str)
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out

df['tagging'] = tagging(df['lemmatized'])
df['tagging'][:5]
```

The results can be visualized in multiple different ways. I've decided to try both a word cloud and a pie chart, due to the many different words, I've decided to limit the pie chart to the 20 most commonly used verbs.

```python


from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Flatten the list of lists of lemmatized words into a single list
all_words = [word for sublist in df['tagging'] for word in sublist]

# Create a string of all words
text = ' '.join(all_words)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the generated image:
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
