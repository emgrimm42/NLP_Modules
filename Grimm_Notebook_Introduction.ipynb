##Variable Assignment

A variable acts like a container or label for a piece of data, allowing you to easily access and manipulate that information later.

When you connect a name to a specific piece of data, you can then use that name to retrieve the data or perform operations on it.

To create a variable, use the = symbol. Put the name you want to give your data on the left side of the equals sign, and the actual data (or "value") on the right side.


*chosen_name = data_or_value*

This action doesn't produce any immediate visible result. Instead, it saves the connection between the name and the data in the computer's memory. You can then use this name throughout your coding session to refer back to that specific piece of data.

You love traveling to different continents. Perhaps you want to write a piece of code to announce your passion.

First, let's describe everything we want the code to do. We could:

2. For each contintent in the list,

```python
SevenContinents = ['Asia', 'Africa', 'North America', 'South America', 'Antarctica', 'Europe', 'Australia']

for continent in SevenContinents:
    print(continent, "is one of the seven continents")
    print("I love", continent)
print("But aren't there only six continents?")

#(Run me!)
```

####Exercise 4: Let's understand what this code is doint using the *re package* and let's improve the code to find ALL dates

The re package handles regular expressions - patterns that find, match, or replace specific text formats in strings, like extracting all email addresses from text with re.findall(r'\S+@\S+', text).

The below regex pattern breaks down as:

*   ([A-Za-z]+) - captures month name (one or more letters)
*   \s+ - matches one or more spaces
*   (\d{1,2}) - captures day (1 or 2 digits)
*   (?:st|nd|rd|th)? - optionally matches ordinal suffixes but doesn't capture them
*   ,? - optionally matches a comma
*   \s+ - matches spaces again
*   (\d{4}) - captures 4-digit year


```python
import re
# Warum steht unten October, 1848?

# Sample text from a historical document
document = """
On March 15, 1848, the revolution began in Vienna.
By April 2nd, 1848, the protests had spread to Budapest.
The emperor fled on May 17, 1848.
Peace was restored in October 1849.
"""

# Pattern to find dates in format: Month Day, Year
month_day_year_date_pattern = r'([A-Za-z]+)\s+(\d{1,2})(?:st|nd|rd|th)?,?\s+(\d{4})'
month_year_date_pattern = r'([A-Za-z]+)\s+(\d{4})'


# Find all dates
full_dates = re.findall(month_day_year_date_pattern, document)
# add code
incomplete_dates = re.findall(month_year_date_pattern, document)
print(incomplete_dates)

print("Dates found in document:")
for month, day, year in full_dates:
    print(f"{month} {day}, {year}")

# add code
for month, year in incomplete_dates:
  print(f"{month}, {year}")


# Count events by year
years = [year for _, _, year in full_dates]+[year for _, year in incomplete_dates]
print(f"\nEvents in 1848: {years.count('1848')}")
print(f"Events in 1849: {years.count('1849')}")
```

###5. Get images from iiif archives

For example, a postcard collection from the Austrian National library

```python
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO

# Get collection
collection_url = 'https://iiif.onb.ac.at/presentation/collection/labs_postcards_all_cards'
collection = requests.get(collection_url).json()

# Get first manifest
manifest_url = collection['manifests'][0]['@id']
manifest = requests.get(manifest_url).json()

# Get image service URL
canvas = manifest['sequences'][0]['canvases'][0]
service_id = canvas['images'][0]['resource']['service']['@id']

# Get image
image_url = f"{service_id}/full/800,/0/default.jpg"
img_data = requests.get(image_url).content
img = Image.open(BytesIO(img_data))

# Show image
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.axis('off')
plt.title(f"Austrian National Library Postcard: {manifest.get('label', 'Historical Postcard')}")
plt.tight_layout()
plt.show()

print(f"Showing: {manifest.get('label', 'Unknown location')}")
```
