## Exercise 3:

Create a new DataFrame where each fulltext contains a specific word. For example "Katastrophe".

```python

#add your code here:

# 1. neue Variable in die nur die Zeilen kommen in denen Katastrophe im der Spalte plainpagefulltext vorkommen
# na = False, alle zeilen ohne Inhalt werden automatisch zu Falsch sortiert
df_katastrophe = df[df['plainpagefulltext'].str.contains('Katastrophe', na = False)]

# 2. Die obersten fünf Zeilen anzeigen lassen
df_katastrophe.head()
```

```python
# 3. zählen wie viele Zeilen es gibt
print(f"Es gibt {len(df_katastrophe)} Zeitungen in denen das Wort Katastrophe vorkommt")
```

```python
# Schaffe ich es die Publikationsorte von df_katastrophe in einem pie chart dar zu stellen?

# Import modules
import pandas as pd
import matplotlib.pyplot as plt

# 1. Publikationsorte raus sortieren
df_katastrophe_pub_ort = df_katastrophe['place_of_distribution']

# Create a pandas Series from df_katastrophe_pub_ort
series = pd.Series(df_katastrophe_pub_ort)

# Count the occurrences of each unique value
# This aggregates the data so the pie chart shows the distribution of skill levels
counts = series.value_counts()

# Display the raw counts
# alter the description
print("Counts of each publication place use:")
print(counts)

# Create a pie chart
plt.figure(figsize=(3, 3))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Verteilung der Publikation von Zeitungen die Katastrophe erwähnen')
plt.show()
```

```python
# Anscheinend haben wir sehr viele Publikationsorte die nicht sonderlich oft vorkommen
# counts, aus der vorherigen box beinhaltet bereits eine Zählung der 78 Publikationsorte mit ihrer Häufung
# schaffen einer neuen Variable mit den top 20
top_20  = counts.head(20)

# Create a pie chart
plt.figure(figsize=(3, 3))
plt.pie(top_20, labels=top_20.index, autopct='%1.1f%%')
plt.title('Verteilung der Publikation von top 20 Zeitungen die Katastrophe erwähnen')
plt.show()
```
